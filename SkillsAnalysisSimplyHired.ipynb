```python
df = pd.read_csv('simplyhired-DataEngineer-2022.csv')
df.head(5)
```

```python
df= df.drop(columns=['Unnamed: 0'])
df = df.drop_duplicates(ignore_index=True)
df.head()
```

```python
df.info()
```

```python
skills_df = df[['qualifications']]
skills_df.info()
```

```python
skills_df.isnull().values.any()
```

```python
skills_df.dropna(subset=["qualifications"],inplace =True)
skills_df.info()
```

```python
skills_df["qualifications"] = skills_df['qualifications'].str.lower().str.replace('[\n]',',')
skills_df
```

```python
result = skills_df.qualifications.str.split(',',expand=True).stack().value_counts().reset_index()
result.columns = ['Word','Frequency']
result.head(20)
```
