## Data cleaning
###  Checking and dropping any duplicate rows
All datasets are currently having first column Unmaed: 0 which is just the row number. This column is not needed and also it is the reason we are not able to delete any duplicate rows. So that column needs to be dropped from all 3 dataframes.

```python
df_dice = pd.concat([df_da,df_de,df_ds])
df_dice= df_dice.drop(columns=['Unnamed: 0'])
df_dice = df_dice.drop_duplicates(ignore_index=True)
df_dice.info()
```

```python
df_dice['title'] = df_dice['title'].str.lower()
df_dice['company'] = df_dice['company'].str.lower()
df_dice['description'] = df_dice['description'].str.lower()
df_dice['location'] = df_dice['location'].str.lower()
df_dice
```

```python
import numpy as np
df_dice['is_remote'] = np.where(df_dice['location'].str.contains('remote'),1,0)
df_dice
```

```python
df_dice['location'] = df_dice['location'].str.replace('d.c.', 'dc')
df_dice['location'] = df_dice['location'].str.strip()
df_dice['location'] = df_dice['location'].str.split(',')
```

```python
df_dice['city'] = df_dice['location'].str[0]
df_dice['state'] = df_dice['location'].str[1]
df_dice['country'] = df_dice['location'].str[2]
df_dice
```

```python
df_dice['city'] = df_dice['city'].str.replace('remote', '')
df_dice['city'] = df_dice['city'].str.replace('remote or  ', '')
df_dice['city'] = df_dice['city'].str.replace(' or ', '')
```

```python
df_dice.info()
```

```python
df_dice.drop(index = [878,1175,2268,2824,2999,511,628,338,947,1743,3041,2076], axis=0,inplace = True)
```

```python
df_dice.country.value_counts()
```

```python
df_dice['country'] = df_dice['country'].str.replace('dc', 'usa')
df_dice.country.value_counts()
```

```python
df_dice['job_type'] = np.where(df_dice['title'].str.contains('analy'),'Data Analyst','0')
df_dice['job_type'] = np.where(df_dice['title'].str.contains('engineer'),'Data Engineer',df_dice['job_type'])
df_dice['job_type'] = np.where(df_dice['title'].str.contains('scien'),'Data Scientist',df_dice['job_type'])
df_dice['job_type'] = np.where(df_dice['job_type'] == '0','Other',df_dice['job_type'])
df_dice
```

```python
experience_levels = ['associate','senior','intern','staff','principal',
                  'director','leader','manager','specialist','entry level','machine learning'
                    ,'junior','jr','sr']

df_dice['experience_level'] = '0'
for index,level in enumerate(experience_levels):
    df_dice['experience_level'] = np.where(df_dice['description'].str.contains(level),level,df_dice['experience_level'])
```

```python
df_dice.experience_level.value_counts()
```

```python
skills=['python','java','sql','spark','hadoop','machine learning','hive','pig','git','nosql','aws','azure', 'power bi', 'tableau','etl','agile','relational databases','postgres','matlab','graphql','data analysis','excel','data modeling','shell scripting','gcp','cloud computing','deep learning','nlp','flask','eda']

df_dice['skills'] = np.nan

for ind in df_dice.index:
    skillSummary=df_dice['description'][ind]
 
    skillSummary = re.sub("[^a-zA-Z]",  # Search for all non-letters
                          " ",          # Replace all non-letters with spaces
                          str(skillSummary))
    skillsArray = ''
    for skill in skills:
        if  re.search(skill,skillSummary):
            skillsArray=skillsArray+skill+','
    skillsArray = skillsArray[:-1]
    #print(skillsArray)
    #print(ind)
    if ind < len(df_dice):
        df_dice['skills'].iloc[ind]=skillsArray
#df.loc[df['skills'].str.contains('NaN',na=True), 'skills'] ='Nothing_found'
df_dice.head(15)

```
